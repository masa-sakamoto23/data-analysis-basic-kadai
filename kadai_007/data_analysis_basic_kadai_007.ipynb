```python
# !pip install japanize_matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

# サンプルcsvファイルを読み込む
df = pd.read_csv('sample_pandas_6.csv')
# df.head()

# カテゴリーcsvファイルを読み込む
category_df = pd.read_csv('category.csv')
# category_df

# dfから「カテゴリー」列を削除してからマージする
df = df.drop(columns=['カテゴリー'], errors='ignore')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], on="商品番号", how="left")
# df

# データフレームのカテゴリー要素の出現頻度をカウントして変数に代入
category_counts = df['カテゴリー'].value_counts()

# シリーズの棒グラフを作成
category_counts.plot(kind='bar')
plt.xlabel('カテゴリー')
plt.ylabel('出現頻度')
plt.title('カテゴリー要素ごとの出現頻度')
plt.show()

# 商品番号列ごとにデータをグループ化して注文数の基本統計量を算出
grouped_describe = df.groupby('商品番号')['注文数'].describe()
print(grouped_describe)
```
