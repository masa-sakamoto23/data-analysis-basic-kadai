```python
import matplotlib.pyplot as plt
import numpy as np

# 
y0 = [1, 2, -5, 2]

# sin
# 010100
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

# 
# 15, 52000
x2 = np.random.normal(15, 5, 2000)

# 
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

# 
x4 = np.array([30, 40, 30])
label4 = ["dogs", "cats", "etc."]

# 
x5 = np.array([1, 2, 3 ,4 ,5])
y5 = np.array([100, 120, 150, 200, 160])
label5 = ["Kelp", "Plum" , "Salmon", "Kalbi", "Sujiko"]


# 
fig, axes = plt.subplots(2, 3, tight_layout=True)

# 
axes[0,0].plot(y0)
# sin
axes[0,1].plot(x1, y1, linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
              ylim=(-1, 5), yticks=np.arange(-1, 6))
# 
axes[0,2].hist(x2, bins=10)
# 
axes[1,0].scatter(x3, y3)
# 
axes[1,1].set_title("Favorite Animals")
axes[1,1].pie(x4, labels=label4)
# 
axes[1,2].bar(x5, y5, tick_label=label5)
axes[1,2].set_title("Onigiri fillings and prices")
# axes[1,2].set_ticks(len(label5))
axes[1,2].set_xticklabels(label5, fontsize=6)

plt.show()
```
